```python
import pandas as pd


df = pd.read_csv("../gisette_train.data", header=None, sep=" ")
df
```

```python
#last column seems to be useless
df = df.iloc[:, :-1]
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standard scale & run PCA with two components
x = StandardScaler().fit_transform(df)
pca = PCA(n_components = 2)
pca_components = pca.fit_transform(x)
pca_df = pd.DataFrame(data=pca_components, columns=["PC1", "PC2"])
pca.explained_variance_ratio_
```

```python
# The top principal component explains 2.88% of variance, and the second top explains 2.26% of variance

import matplotlib.pyplot as plt

%matplotlib inline

labels = pd.read_csv("../gisette_train.labels", header=None, sep=" ")

plt.figure()
plt.xlabel("PC1")
plt.ylabel("PC2")

targets = [1, -1]
colors = ["b", "r"]

for target, color in zip(targets, colors):
        indToKeep = labels[0] == target
        plt.scatter(pca_df.loc[indToKeep, "PC1"],
                    pca_df.loc[indToKeep, "PC2"],
                    c=color)
```

```python
# All principal components -> explained_variance_ratio

pca_2 = PCA().fit(x)
evr = pca_2.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

```python
import numpy as np

plt.plot(np.cumsum(pca_2.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance')
```
